```python
%%zipline --start 2017-11-1 --end 2018-1-1 -o dma.pickle

from zipline.pipeline.factors import Returns
from zipline.pipeline import Pipeline
from zipline.api import attach_pipeline, pipeline_output
from zipline.pipeline.fundamentals import Fundamentals
from zipline.pipeline.data import CNEquityPricing

def make_pipeline():
    recent_returns = Returns(
        window_length=5, 
    )
    
    # Turn our recent_returns factor into a z-score factor to normalize the results.
    recent_returns_zscore = recent_returns.zscore()
    
    pipe = Pipeline(
        columns={
            'recent_returns_zscore': recent_returns_zscore,
            'close':CNEquityPricing.close.latest,
        },
    )
    return pipe

def initialize(context):
    attach_pipeline(make_pipeline(), 'example')

def handle_data(context, data):
    # Pipeline_output returns the constructed dataframe.
    output = pipeline_output('example')

    print(output.tail())
```

```python
%%zipline --start 2017-11-1 --end 2018-1-1 -o dma.pickle

from zipline.pipeline.factors import Returns
from zipline.pipeline import Pipeline
from zipline.api import attach_pipeline, pipeline_output

from zipline.pipeline.fundamentals import Fundamentals

def make_pipeline():
    cap = Fundamentals.info.注册资本.latest
    ind = Fundamentals.info.国证四级行业.latest
    d = Fundamentals.info.上市日期.latest
    recent_returns = Returns(
        window_length=5, 
    )
    
    # Turn our recent_returns factor into a z-score factor to normalize the results.
    recent_returns_zscore = recent_returns.zscore()
    
    return Pipeline(
        columns={
            '注册资本': cap,
            '国证四级行业':ind,
            '上市日期':d,
            'recent_returns_zscore': recent_returns_zscore,
            '流通股本':Fundamentals.equity.流通股本.latest,
        },
    )

def initialize(context):
    attach_pipeline(make_pipeline(), 'example')

def handle_data(context, data):
    # Pipeline_output returns the constructed dataframe.
    output = pipeline_output('example')

    print(output.tail())
```
